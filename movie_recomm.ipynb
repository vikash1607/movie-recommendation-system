```python
import numpy as np
import pandas as pd
```

```python
credits=pd.read_csv('tmdb_5000_credits.csv')
movies=pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
#random 5 data rows and columns
credits.sample(5)
```

```python
#merging both data 
movies=movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
movies.head()
```

```python
#genre , id ,keywords , title , overview , cast , crew there columns will be selected for better output
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
#checking null values
movies.isnull().sum()
```

```python
#dropping null values wrt columns
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
#checking duplicate row
movies.duplicated().sum()
```

```python
#these given are list of string , so convert into list [ast.literal_eval()]
import ast 
```

```python
movies.iloc[0].genres
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#take first three  values out of cast only
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#take director values out of crew only
def fetch_dir(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
movies.head()
```

```python
#convert string into list to concatenate with all other columns
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove space between keywords for all other columns like Vikash Kumar will convert into VikashKumar to avoid error
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
#concatenate all given columns 
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies['tag'][0]
```

```python
new_df=movies[['movie_id','title','tag']]
new_df
```

```python
#remove all , from the tag column 
def space(lst): 
    return ' '.join(lst)
```

```python
new_df['tag']=new_df['tag'].apply(space)
```

```python
new_df.head()
```

```python
#convert tag columns into lowercase
new_df['tag']=new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df['tag']
```

```python
new_df['tag'][2]
```

```python
new_df.head()['tag'][0]
```

```python
# convert words of similar to single keywords like [loving,loved,loves] to [love]
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
# convert words of similar to single keywords like [loving,loved,loves] to [love]
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
new_df.head()['tag'][0]
```

```python
#converting text to vectors (bag of words technique)
#finding similiarity between tags using library and removing english words , max 5000 words allowed and will skip on,in,the words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#convertd to array with no of repeated words
vectors=cv.fit_transform(new_df['tag']).toarray()
```

```python
vectors[0]
```

```python
#cv.get_feature_names()
```

```python
#in higher dimension we will prefer cosine distance b/w two points  
# finding distance through cosine similarity rather than euclidian distance 
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#an array of vector made from data where distance from each point in done 
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
#important
#first line : find movie index 
#seconf line :  dinf distance from similarity vector
#third lind: sort in decreasing ordrer and sort with resp to second value
#enumerate will hold index with values ,[1:6] i.e 5 top similar values
def reccomend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
#give u similar movie name 
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
new_df[new_df['title']=='Avatar'].index[0]
```

```python
reccomend('Batman Begins')
```
